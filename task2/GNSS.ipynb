```python
import pandas as pd
import numpy as np
from operator import itemgetter, attrgetter
from geopy.distance import geodesic
from geopy import distance
import datetime
import math
from math import radians, cos, sin, asin, sqrt, ceil
from numpy import nan as NaN
import folium
import webbrowser
from folium.plugins import HeatMap
```

```python
use_cols = ['latitude','longitude','determination_position']

print(datetime.datetime.now())
for index in range(1,46):
    df= pd.read_csv("C:\\Users\\qiusu\\JupyterNotebook\\Daten\\Lang_Daten\\{:02}_211203_TUDA_data.csv".format(index),usecols=use_cols)
    df = df[(df['latitude']<=72)&(df['latitude']>=36)&(df['longitude']<=42)&(df['longitude']>=-10)]
    
    df['la_2']=(round(df.latitude*5)/5)
    df['lo_2']=(round(df.longitude*5)/5)
    df['la_1']=(round(df.latitude,1))
    df['lo_1']=(round(df.longitude,1))
    df['la_05']=(round(df.latitude*2,1)/2)
    df['lo_05']=(round(df.longitude*2,1)/2)
    df['la_01']=(round(df.latitude,2))
    df['lo_01']=(round(df.longitude,2))
    
    df['ratio_01']=df.groupby(['la_01','lo_01'],as_index=False)['la_1'].transform('count')
    df['ratio_05']=df.groupby(['la_05','lo_05'],as_index=False)['la_1'].transform('count')
    df['ratio_1']=df.groupby(['la_1','lo_1'],as_index=False)['la_1'].transform('count')
    df['ratio_2']=df.groupby(['la_2','lo_2'],as_index=False)['la_1'].transform('count')

    #df= df[(df['ratio_01']<=1)&(df['ratio_x']>=400)]
    df=df[((df['ratio_01']<=1)&(df['ratio_1']>=2000))|((df['ratio_05']<=1)&(df['ratio_1']<2000)&(df['ratio_1']>=100))|(df['ratio_2']<=1)]
    if index==1:
        df_sum=df
    else:
        df_sum=pd.concat([df_sum,df],ignore_index=True) 
    
print(datetime.datetime.now())
df_sum.to_csv("C:\\Users\\qiusu\\JupyterNotebook\\Daten\\GNSS_Output_ratio_2.csv") 
```
